```
import sklearn
import numpy as np
import pandas as pd
import plotly.express as px

import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 32)
```

```
df_case = pd.read_csv("/work/time_series_covid19_confirmed_global.csv", sep = ',')
```

```
df_case.describe()
```

```
df_case =df_case.drop(['Lat', 'Long'],axis=1)
df_case =df_case.drop(['Province/State'],axis=1)
```

```
df_case = df_case.groupby(['Country/Region']).sum()
```

```
df_case
```

```
df_case = df_case.T.reset_index().reindex()
df_case =df_case.rename(columns = {'index':'Date'})
```

```
df_case['index'] = pd.to_datetime(df_case['index']).dt.date
```

```
df = px.data.gapminder().query("continent=='Oceania'")
fig = px.line(df_case, x="Date", y=df_case.columns)
fig.show()
```

```
#Deaths
df_deaths = pd.read_csv("/wor/work/time_series_covid19_deaths_global.csv", sep = ',')
df_deaths =df_deaths.drop(['Lat', 'Long'],axis=1)
df_deaths =df_deaths.drop(['Province/State'],axis=1)
df_deaths = df_deaths.groupby(['Country/Region']).sum()
df_deaths = df_deaths.T.reset_index().reindex()
df_deaths =df_deaths.rename(columns = {'index':'Date'})
df_deaths['Date'] = pd.to_datetime(df_deaths['Date']).dt.date
df_deaths = px.data.gapminder().query("continent=='Oceania'")
fig = px.line(df_deaths, x="pop", y=df_deaths.columns)
fig.show()
```
